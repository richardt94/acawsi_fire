```python
import ipyleaflet as L
```

```python
from ipywidgets import Label, Layout, VBox

label = Label(layout=Layout(width='100%'))

map_catchments = L.Map(center=(-30,130),zoom=3)

catch_layer = L.GeoData(geo_dataframe = catchments,
                   style={'color': 'black', 'fillColor': 'red', 'opacity':0.05, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6},
                   hover_style={'fillColor': 'red' , 'fillOpacity': 0.2},
                   name = 'Countries')

def catch_hover(event=None, feature=None, id=None, properties=None):
    label.value = properties['TAG']
    
catch_layer.on_hover(catch_hover)

river_layer = L.GeoData(geo_dataframe = rivers,
                   style={'color': 'black', 'fillColor': '#3366cc', 'opacity':0.05, 'weight':1.9, 'dashArray':'2', 'fillOpacity':0.6},
                   hover_style={'fillColor': 'red' , 'fillOpacity': 0.2},
                   name = 'Countries')

map_catchments.add_layer(catch_layer)
map_catchments.add_layer(river_layer)

VBox([map_catchments,label])
```

```python
from affine import Affine
transform = Affine.translation(112,-10)*Affine.scale(0.05,-0.05)
```

```python
from rasterio import features
```

```python
catch_geom = catch_geom.iloc[0]
```

```python
catch_mask = features.geometry_mask([catch_geom],awra_monthly.qtot.shape,transform,invert=True)
```

```python
import matplotlib.pyplot as plt
plt.imshow(catch_mask.astype(float))
```

```python
awra_nsw_qtot = awra_monthly.where(catch_mask).qtot
```

```python
awra_nsw_qtot.dropna(dim='latitude',how='all').dropna(dim='longitude',how='all').plot(robust=True)
```
