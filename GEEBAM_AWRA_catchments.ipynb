## Open GEEBAM GeoTIFF

```python
GEEBAM = xr.open_rasterio('/g/data/oe9/user/rlt118/GEEBAM_v3p0_dNBR_Classes.tif')
```

```python
#resample GEEBAM to be a manageable resolution
import numpy as np
GEEBAM_lr = GEEBAM.reindex(
        x = np.arange(GEEBAM.x.min(),GEEBAM.x.max(),0.005),
        y = np.arange(GEEBAM.y.max(),GEEBAM.y.min(),-0.005),
        method = 'nearest',
        tolerance = GEEBAM.res[0]   
)
```

```python
GEEBAM_lr = GEEBAM_lr.isel(band=0)
```

```python
GEENSW = GEEBAM_lr.sel(x=slice(nsw.bounds[0],nsw.bounds[2]),y=slice(nsw.bounds[3],nsw.bounds[1]))
```

```python
nsw.bounds
```

```python
GEENSW.plot()
```

```python
GEENSWtrans = Affine.translation(GEENSW.x[0].data,GEENSW.y[0].data)*Affine.scale(0.005,-0.005)
```

```python
gmask = features.geometry_mask([nsw],GEENSW.shape,GEENSWtrans,invert=True)
```

```python
GEENSW = GEENSW.where(gmask)
```

```python
GEENSW.plot()
```

```python
import pandas as pd
tiffmap = pd.Series(GEEBAM_class_table.GEEBAMclas.values,index=GEEBAM_class_table.Value).to_dict()
tiffmap[253] = 253
```

```python
tiffmap[253]
```

```python
severity_dict = {
    0 : 3,
    1 : 1,
    2 : 2,
    3 : 3,
    4 : 4,
    253 : 0,
}

@np.vectorize
def tiffmap_float(val):
    if np.isnan(val):
        return val
    idx = int(val)
    sev_cat = tiffmap[idx] if tiffmap[idx] != 254 else 0
    return severity_dict[sev_cat]
    
    
burntNSW = xr.apply_ufunc(tiffmap_float,GEENSW)
```

```python
burntNSW.plot()
```

```python
burntNSW = burntNSW.rename({'x':'longitude','y':'latitude'})
```

```python
awra_nsw_qtot = awra_nsw_qtot.reindex(longitude=burntNSW.longitude,latitude=burntNSW.latitude,method='nearest',tolerance=0.05)
```

```python
risk_map = awra_nsw_qtot*burntNSW
risk_map.plot(vmin=0,vmax=250,figsize=(15,11))
```

```python
risk_map.quantile(0.95)
```

```python
len(catchments)
```

```python
rmap_transform = Affine.translation(risk_map.longitude.min(),risk_map.latitude.max())*Affine.scale(0.005,-0.005)
```

```python
def catch_to_risk(catchment_poly):
    #mask the risk map to the shape of the catchment
    mask = features.geometry_mask([catchment_poly],risk_map.shape,rmap_transform,invert=True)
    mean_risk = risk_map.where(mask).mean()
    quantiles = risk_map.where(mask).quantile([0.5,0.95])
    
    result = np.array([mean_risk,*quantiles.data])
    result[np.isnan(result)] = 0
    return result
```

```python
catch_to_risk(catchments.iloc[43].geometry)
```

```python
risk_results = np.stack(catchments.geometry.apply(catch_to_risk))
```

```python
catchments['mean_risk'] = risk_results[:,0]
catchments['median_risk'] = risk_results[:,1]
catchments['p95_risk'] = risk_results[:,2]
```
